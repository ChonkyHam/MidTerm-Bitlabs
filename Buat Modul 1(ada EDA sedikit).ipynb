```python
numerical = [var for var in df.columns if df[var].dtype!='O']
```

```python
import numpy as np
L = len(numerical)

# Define number of columns and rows for the subplots
ncol = 3
nrow = int(np.ceil(L / ncol))

# Create a figure with subplots
fig, axes = plt.subplots(nrow, ncol, figsize=(15, 5 * nrow))
fig.subplots_adjust(top=0.92)

# Flatten the axes array for easier iteration
axes = axes.flatten()

# Loop through numerical columns and plot
for i, col in enumerate(numerical):
    sns.histplot(data=df, x=col, hue="is_canceled", multiple="stack", ax=axes[i], bins=30, palette="coolwarm")
    axes[i].set_title(f'Distribution of {col}', fontsize=14)
    axes[i].set_xlabel(f'{col} Value', fontsize=12)
    axes[i].set_ylabel('Frequency', fontsize=12)
    axes[i].grid(True, which='both', linestyle='--', linewidth=0.5)
    sns.despine(ax=axes[i])


# Turn off any extra subplots (if the number of numerical features is less than nrow*ncol)
for j in range(L, nrow * ncol):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Boxplot for lead_time based on cancellation status
plt.figure(figsize=(12, 6))

# Plot 1: Boxplot
plt.subplot(1, 2, 1)
sns.boxplot(x='is_canceled', y='lead_time', data=df)
plt.title('Distribusi Lead Time Berdasarkan Pembatalan')
plt.xticks([0, 1], ['Tidak Dibatalkan', 'Dibatalkan'])
plt.xlabel('Status Pembatalan')
plt.ylabel('Lead Time')

# Plot 2: Histogram
canceled = df[df['is_canceled'] == 1]['lead_time']
not_canceled = df[df['is_canceled'] == 0]['lead_time']
mean_canceled = canceled.mean()
mean_not_canceled = not_canceled.mean()

plt.subplot(1, 2, 2)
sns.histplot(canceled, label="Dibatalkan", color="r", kde=True)
sns.histplot(not_canceled, label="Tidak Dibatalkan", color="g", kde=True)
plt.axvline(mean_canceled, color='r', linestyle='--', label=f"Mean Canceled: {mean_canceled:.2f}")
plt.axvline(mean_not_canceled, color='g', linestyle='--', label=f"Mean Not Canceled: {mean_not_canceled:.2f}")
plt.title('Distribusi Lead Time')
plt.xlabel('Lead Time')
plt.ylabel('Jumlah Reservasi')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Nomor 4

# Filter untuk pemesanan last minute
last_minute_bookings = df[df['lead_time'] == 0 ]

# Hitung total pemesanan last minute
total_last_minute_bookings = last_minute_bookings['lead_time'].count()

# Hitung total pembatalan pemesanan last minute
last_minute_cancellations = last_minute_bookings[last_minute_bookings['is_canceled'] == 1]
total_last_minute_cancellations = last_minute_cancellations['lead_time'].count()

# Hitung tingkat pembatalan pemesanan last minute
if total_last_minute_bookings > 0:
    cancellation_rate = (total_last_minute_cancellations / total_last_minute_bookings) * 100
else:
    cancellation_rate = 0

# Cetak hasil analisis
print(f"Total Last-Minute Bookings: {total_last_minute_bookings}")

print(f"Total Last-Minute Cancellations: {total_last_minute_cancellations}")
print(f"Cancellation Rate for Last-Minute Bookings: {cancellation_rate:.2f}%")

# Analisis karakteristik pemesanan last minute
plt.figure(figsize=(10, 6))
sns.histplot(last_minute_bookings['adr'], bins=20, kde=True)
plt.title('Distribusi Harga Pemesanan Last Minute')
plt.xlabel('Harga per Malam (ADR)')
plt.ylabel('Frekuensi')
plt.show()
```
