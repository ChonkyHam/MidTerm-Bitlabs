```python
# Menghubungkan Google Drive dan membaca file midterm_hotel_data ke dalam DataFrame
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd

file_path = '/content/drive/MyDrive/MSIB Batch 7/Hands on Assignment/MidTerm/midterm_hotel_data.csv'
df = pd.read_csv(file_path)
```

```python
# Baris kode ini untuk menghapus semua duplikat
df.drop_duplicates()

# df = df.drop(['Unnamed: 0'], axis=1) -> Menghapus Kolom Unnamed: 0 yang tidak berguna

# Bagian kode ini untuk mengisi data yang awalnya kosong menjadi 0
df['lead_time'] = df['lead_time'].fillna(0)
df['stays_in_weekend_nights'] = df['stays_in_weekend_nights'].fillna(0)
df['adults'] = df['adults'].fillna(0)
df['children'] = df['children'].fillna(0)
df['country'] = df['country'].fillna(0)
df['agent'] = df['agent'].fillna(0)
df['company'] = df['company'].fillna(0)
df['adr'] = df['adr'].fillna(0)
df['total_of_special_requests'] = df['total_of_special_requests'].fillna(0)

# Baris kode ini untuk menampilkan tabel
display(df)

# Baris kode ini untuk membuat file baru bernama "cleaned_data" dimana datanya sudah bersih
df.to_csv('/content/drive/MyDrive/Studi Independen MSIB/Hands on Assignment/MidTerm/cleaned_data.csv', index=False)
```
