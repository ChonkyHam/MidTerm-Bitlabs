```python
# Menghubungkan Google Drive dan membaca file cleaned_data ke dalam DataFrame
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

file_path = '/content/drive/MyDrive/MSIB Batch 7/Hands on Assignment/MidTerm/cleaned_data.csv'
df = pd.read_csv(file_path)
```

```python
# Memisahkan data ADR untuk kedua kelompok hotel
# Kelompok A: City Hotel
# Kelompok B: Resort Hotel
adr_hotel_city = df[df['hotel'] == 'City Hotel']['adr']
adr_hotel_resort = df[df['hotel'] == 'Resort Hotel']['adr']

# Melakukan A/B Testing menggunakan independent t-test
nilai_t, nilai_p = stats.ttest_ind(adr_hotel_city, adr_hotel_resort, equal_var=False)

# Mencetak hasil pengujian
print("Hasil A/B Testing:")
print("-"*50)
print(f"Statistik Uji (T-statistic): {nilai_t:.2f}")
print(f"Nilai-P (P-value): {nilai_p:.4f}")

# Menetapkan tingkat signifikansi
alpha = 0.05

# Membuat keputusan
print("\nKeputusan:")
print("-"*50)
if nilai_p < alpha:
    print("Karena nilai-p < 0.05, maka MENOLAK H0")
    print("Kesimpulan: Terdapat perbedaan yang signifikan pada rata-rata ADR antara City Hotel dan Resort Hotel")
    print("Hipotesis yang diterima: H1")
else:
    print("Karena nilai-p >= 0.05, maka GAGAL MENOLAK H0")
    print("Kesimpulan: Tidak ada perbedaan yang signifikan pada rata-rata ADR antara City Hotel dan Resort Hotel")
    print("Hipotesis yang diterima: H0")
```
